```python
class Circle:
    def __init__(self, x=0, y=0, radius=0):
        self.x = x
        self.y = y
        self.radius = radius

c1 = Circle(1,1,2)
c2 = Circle(radius = 1)

```

```python
import matplotlib.pyplot as plt 
from matplotlib import patches

fig, ax = plt.subplots(1)

circle_patch1 = patches.Circle((c1.x, c1.y), radius = c1.radius, fill = False)
circle_patch2 = patches.Circle((c2.x, c2.y), radius = c2.radius, fill = False)
ax.add_patch(circle_patch1)
ax.add_patch(circle_patch2)
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
ax.grid()
```
