- in this course, focus on the components of a matplotlib figure 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# generate data x
x = np.linspace(-5, 5, 100)
x
```

```python
# generate data f
f = np.sin(x)
f
```

##### The first approach - functional/matlab style 

```python
plt.figure()
plt.plot(x,f)
plt.title("sin(x)")
plt.xlabel("x")
plt.ylabel("f")
```

##### The second approach - OOP

```python
fig = plt.figure() # create one figure
ax = plt.axes() # on the figure, create an axe
ax.plot(x, f) # on the axe, create a plot
ax.set(xlim=(-10,10), ylim=(-5,5), xlabel="x", ylabel="y") # continue modify the axe object in different aspects

# check out more about the components of matplotlib figure here: https://matplotlib.org/stable/gallery/showcase/anatomy.html (optional reading)
```
